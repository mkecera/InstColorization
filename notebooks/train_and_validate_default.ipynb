## Initialize hyperparameters

```python
DATASET_DIR="./train_tiny"
LEARNING_RATE = 0.0005
EPOCHS = 5
EXPERIMENT_NAME="default"
```

## Prepare trainboxes

```python
!python inference_bbox.py --test_img_dir $DATASET_DIR --filter_no_obj
```

## Train first stage - Full module

```python
!mkdir ./checkpoints/coco_full
!cp ./checkpoints/siggraph_retrained/latest_net_G.pth ./checkpoints/coco_full/
!python train.py --stage full --name coco_full --sample_p 1.0 --niter $EPOCHS --niter_decay 1 --load_model --lr $LEARNING_RATE --model train --fineSize 256 --batch_size 16 --display_ncols 3 --display_freq 1600 --print_freq 1600 --train_img_dir $DATASET_DIR --no_html
```

## Train second stage - Instance module

```python
!mkdir ./checkpoints/coco_instance
!cp ./checkpoints/coco_full/latest_net_G.pth ./checkpoints/coco_instance/
!python train.py --stage instance --name coco_instance --sample_p 1.0 --niter $EPOCHS --niter_decay 1 --load_model --lr $LEARNING_RATE --model train --fineSize 256 --batch_size 16 --display_ncols 3 --display_freq 1600 --print_freq 1600 --train_img_dir $DATASET_DIR --no_html
```

## Train third stage - Fusion module

```python
!mkdir ./checkpoints/$EXPERIMENT_NAME
!cp ./checkpoints/coco_full/latest_net_G.pth ./checkpoints/$EXPERIMENT_NAME/latest_net_GF.pth
!cp ./checkpoints/coco_instance/latest_net_G.pth ./checkpoints/$EXPERIMENT_NAME/latest_net_G.pth
!cp ./checkpoints/coco_full/latest_net_G.pth ./checkpoints/$EXPERIMENT_NAME/latest_net_GComp.pth
!python train.py --stage fusion --name $EXPERIMENT_NAME  --sample_p 1.0 --niter $EPOCHS --niter_decay 0 --lr $LEARNING_RATE --model train --load_model --display_ncols 4 --fineSize 256 --batch_size 1 --display_freq 1 --print_freq 1 --train_img_dir $DATASET_DIR --no_html
```
